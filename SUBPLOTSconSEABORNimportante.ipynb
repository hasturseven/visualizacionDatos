```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.head(2)
```

## AHORA USANDO LOS SUBPLOTS CON SEABORN QUE ES GRAFICAS MAS DE UNO

```python

fig ,axes = plt.subplots(2,2, figsize=(10,7))
sns.scatterplot(ax=axes[0,0],data=tips,x='total_bill',y='tip',hue='day',style='time');
sns.boxplot(ax=axes[1,1],data= tips, x= 'day', y = 'total_bill',  hue= 'sex');

# y aso sucesibamente con las que yo quiera llenar poniendolas donde yo quiera
```
