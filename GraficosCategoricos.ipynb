# GRAFICOS CATEGORICOS

>seaborn ofrece diferentes gráficos para graficar datos categóricos o variables de texto. A continuación verás el código que podrás usar para gráficar este tipo de datos.


# Tipos de datos categóricos para graficar en Seaborn



Empieza importando los datos que necesitas de las librerías

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips= sns.load_dataset("tips")
tips.head(3)
```

# Gráfica de cantidad

la mas sencilla pero una de las mas utiles
>estamos ocntando la variable categorica osea cuantas propinas segun el dia y estara separado por el gnero sexx

```python
sns.countplot(data=tips,x='day',hue='sex');
```

>ahora poniendolo en el eje y

```python
sns.countplot(data=tips,y='day',hue='sex')
```

# para varaibles categoricas tambien tenemos stippplot Diagrama de puntos

> aca miraremos el total de la cuenta

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex');
```

>ahora separaremos las variables como siempre usando dodge=true
>usaremos el plt para retocar un poco la imagen osea para verla mas grande

```python
plt.figure(figsize=(14,9))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True);
```

> Otro Diagrama de puntos es el swarmplot
> > sino que en su cncentracion de datos tambien tendremos un eje horizontal que nos da como un poco mas de info , investigarlo un poco mas que el de platzi explica una mierda

```python
plt.figure(figsize=(12, 7))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True);
```

# Diagrama de caja

```python
plt.figure(figsize=(8,7))
sns.boxplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True );
```

# MUY IMPORTANTE TAMBIEN PUEDO CONBINAR GRAFICAS ENTRE SI

```python
plt.figure(figsize=(8,7))
sns.boxplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True);
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True,color=(0.11, 0.420, 0.15, 0.8),marker= '<');
```

# Violín: parecido a Boxplot, pero no muestra los cuartiles, sino como se concentran los datos con comparación split


```python
plt.figure(figsize=(8,7))
sns.violinplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', split=True);
#el split es para unirlos si no se usa quedan separados como si usaramos el dodge
```

>> tambien lo podemos dividir con el doge

```python
plt.figure(figsize=(8,7))
sns.violinplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True);
```

# muy importante Catplot, con kind podemos cambiar el tipo de gráfico y se puede agregar otra variable col
DE LOS MEJORES PARA TRABAJAR
>el col es importante como hay distintas variables crea distintas graficas para cada uno como pasa a la hora del almuerxo y la cena


```python
sns.catplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True, kind= 'box', col= 'time');
```

```python
sns.catplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True, kind= 'swarm', col= 'time');
```
