```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = np.sin(x)
```

Crear una gr치fica tipo Fig

>Para crear una gr치fica que sea tipo Fig. necesitas trabajar el siguiente c칩digo:


```python
fig, axes = plt.subplots()

```

#AL HACER EL FIG,AXES = PLT.SUBPLOTS()
LE ESTAMOS diciendo creame una figura y axes es como decir el nombre de la grafica mas adelante vemos que podemos tener mas de una grafica dentro del fig que es como el luenzo o el macro y para identificar cada grafica usamos axes[0] si como una matriz

AHORA le estamos pasando los valores para que me grafique correctamente la imagen

```python
fig, axes = plt.subplots()
axes.plot(x,y,'y')
plt.show()
```

AHORA CREANDO VARIAS GRAFICAS O LIENZOS PARA PODER GRAFICAR
>VARIAS GRAFICAS TIPO FIG

AL HACER USO DE LA N ROWS ES DECIR CUANTAS FILAS Y CUANTAS COUMNAS Y ME CREARA VARAIS GRAFICAS DEPENDIENDO LO QUE LE MANDE

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
```

AHORA COMO TENEMOS UNA FILA Y DOS COLUMNAS EL AXES HACE REFERENCIA A LOS GRAFICOS DENTRO DE LA FIGURA ENTONCES COMO ESTA EN FORMA DE ARRAY USAMOS LOS INDICES

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'y')
axes[1].plot(y,x,'r')
plt.show()
```

OTRA MANERA DE TRABAJAR ESTE TIPO DE GRAFICAS QUE SON MAS DE UNA QUE SE PRESENTARAN COMO ARRAYS , PUES PODEMOS USAR UNA TUPLA Y DARLE EL NOMBRE PARA CADA GRAFICA

```python
fig, (grafica1,grafica2) = plt.subplots(nrows=1,ncols=2)
grafica1.plot(x,y,'y')
grafica2.plot(y,x,'r')
plt.show()
```

EN ESTE CASO USAREMOS DOS FILAS CUATRO COLUMNAS PARA ACEDER A CADA GRAFICO SE HACE CON LA POSICION COMO EN LAS MATRICES
y aquellos que no le pasemos valores los dejara por defecto

PARA MEJORAR LA PRESENTACION USAMOS UNA PROPIEDAD DEL OBJETO TIPO FIG PARA QUE SE VEA MEJOR

```python
fig, axes = plt.subplots(nrows=2,ncols=4)
axes[0,0].plot(x,y,'y')
axes[0,1].plot(x,np.cos(x),'r')
axes[0,2].plot(x,np.tan(x),'b')
axes[0,3].plot(x,np.tan(y),'c')
axes[1,1].plot(x,np.sin(y),'g')
axes[1,0].plot(y,x,'r')
fig.tight_layout()
plt.show()
```
