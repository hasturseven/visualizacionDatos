```python
import matplotlib.pyplot as plt
import numpy as np
```

GRAFICA DE BARRAS :
El gráfico de barras nos permite graficar variables categóricas, es decir, variables de texto, lo que es muy importante en el mundo de la ciencia de datos y Matplotlib ofrece ciertas características que nos facilita la vida en cuanto a graficar este tipo de variables.

>Cómo empezar a crear Bar Plot
>> USO DE VARIABLES CATEGORICAS

```python
country = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [5000, 800, 900, 1000, 300]
```

ahora usando las variables de letras y la numericas generaremos el graficado de barras
usando el plt.bar que es el grafico de tipo barra le pasaos primero los datos del eje x, y luego eje y

```python
plt.bar(country,population)
plt.ylabel('cantidad de poblacion')
plt.xlabel('paises ')
plt.show()
```

>MODIFICANDO COLORES Y TAMAÑOS
>>le pasamos el color en una lista para que le de ese color segun el orden especifico de mis paises

```python
plt.bar(country,population,  width=0.5, color= ["aqua", "grey", "teal", "yellow", "red"])
plt.ylabel('cantidad de poblacion',color='red')
plt.title('paises vs personas')
plt.grid(True) #el plt grid es para que me cree esa grilla que tengo
plt.xlabel('paises ')
plt.show()
```

SI LE AGREGAMOS DOS COLORES NO MAS LOS ALTERNARA

```python
plt.bar(country,population,  width=0.5, color= ["aqua", "YELLOW"])
plt.ylabel('cantidad de poblacion')
plt.xlabel('paises ')
plt.show()
```

>Modificar los nombres de las variables en el eje ‘X’ con xticks
> >esto pasa cuando no puedo controlar los nombres de las variables del eje x entonces me toca hacer uso del xtick para arle el nombre bien

```python
plt.bar(country,population,  width=0.5, color= ["aqua", "YELLOW"])
plt.ylabel('cantidad de poblacion')
plt.xlabel('paises ')
plt.xticks(np.arange(len(country)), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'))
plt.show()
```

>ahora vamos a girar los nombres en el eje x

```python
plt.bar(country,population,  width=0.5, color= ["aqua", "YELLOW"])
plt.xticks(np.arange(len(country)), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'),rotation=46)
plt.grid(True) #el plt grid es para que me cree esa grilla que tengo
plt.ylabel('cantidad de poblacion')
plt.xlabel('paises ')
plt.show()
```

>Crear un gráfico de manera horizontal con barh

```python
plt.barh(country,population,color=['green','yellow'])
plt.xlabel('cantidad de poblacion')
plt.grid(True) #el plt grid es para que me cree esa grilla que tengo
plt.ylabel('paises ')
plt.yticks(rotation=67)
plt.show()
```
