PARA QUE NUESTROS GRAFICOS TENGAN SENTIDO NECESITAMOS  dar contexto a nuestros gráficos necesitamos usar títulos, leyendas, tamaño o etiquetas, para que nuestra gráfica tenga un contexto más amplio.
>Creamos nuestras gráficas comunes


```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,5,11)
y=np.sin(x)

fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'y')
ax2.plot(y,x, 'r')
plt.show()
```

AHORA LE DAREMOS EL CONTEXTO A LOS GRAFICOS COMENZANDO CON LO BASICO EL NOMBRE
figsize pone mas grande la grafica en una escala mayor entonces seleccionanamos un numero para dejarlo mas gande como podemos ver
con el ax1.legend() hace que sea vea label= osea el nombre de la figura

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.plot(x,y, 'y',label="$sin(x)$")
ax1.set_title('relacion X and Y')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.legend()
ax2.plot(y,x, 'r')
ax2.set_title('Relacion Y and X')
ax2.set_xlabel('eje X')
ax2.set_ylabel('eje Y')
plt.show()
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left',bbox_to_anchor=(1,0.3)) #loc es el lugar donde queramos que aparezca el label por defecto la pone en el mejor lugar
#el bbox_to_anchor hace correr la etiqueta a otro lado el ancho etc....
```
