Podemos personalizar mejor nuestros gráficos con diferentes colores y estilos, así, se entenderá mucho mejor nuestras gráficas.
Cómo agregar colores y estilos en Matplotlib

>Estilos de gráficas que hay


```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,5,11)
y=np.sin(x)
```

>Muestra los stilos que tiene matplot

```python
print(plt.style.available)
```

>Usen este mega hack para poder ver todos los estilos aplicados

```python
for i in plt.style.available:
  plt.style.use(i)
  fig, ax = plt.subplots(figsize=(2,2))
  ax.plot(y,x,'y')
  ax.set_title(i, size=20)
  plt.show()
```

SIEMPRE QUE QUERAMOS USAR UN ESTILO EN ESPECIFICO DEBEMOS DEFINIRLO PRIMERO COMO VEMOS EN EL SIG EJEMPLO

>Usar un estilo de grilla solarizado

```python
plt.style.use('Solarize_Light2')
```

```python
fig, ax = plt.subplots(figsize=(2,2))
ax.plot(y,x,'y')
plt.show()
```

AHORA VAMOS A PERSONALIZAR COLORES Y ESTILOS EN PYPLOY
Para personalizar con pyplot es un poco más largo cuando programamos, pero se puede personalizar muchísimo más

>personalizacion del color

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize = (6,6))
ax1.plot(x,x+1,color = '#D426C8') # ---> Color RGB
ax1.plot(x,x+2,color = '#66FF89')

ax2.plot(x,x+3,color = 'blue') #--> Color común
ax2.plot(x,x+4, color = 'red')
```

>Modificando Transparencia (alpha) y grosor (linewidth)

```python
fig, (ax1,ax2) = plt.subplots(2,1,figsize = (6,6))
ax1.plot(x,x+1,color = '#D426C8', alpha= 0.5, linewidth= 18)
ax1.plot(x,x+2,color = '#66FF89', linewidth= 3)
ax2.plot(x,x+3,color = 'blue', linewidth= 5)
ax2.plot(x,x+4, color = 'grey', alpha = 0.3, linewidth= 12)
```

>Estilo de la linea (linestyle) y marcadores (marker)

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = 'X')
ax.plot(x,x+4, color = 'black',linestyle = '--', marker = 'P')
```

>Tamaño de marcadores (markersize) y color (markerfacecolor)

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "#37D842")

ax.plot(x,x+4,color = '#D426C8', alpha= 0.5, linewidth= 18,linestyle = '--', marker = 'P', markerfacecolor="#FF0000")
```
