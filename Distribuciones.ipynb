Seaborn ofrece algunas gráficas u opciones que te permiten trabajar con distribuciones orientadas a tipos de datos numéricos. Veamos como puedes lograrlo utilizando código.

# RECORDAR QUE SIEMPRE ES MEJOR USAR SNS.DISPLOT() QUE ES LA FORMA MAS GENERAR Y HACE LOS MISMO GRAFICOS Y PODEMOS EXPSIFICAR CUAL DE ESTOS QUE VIENE EN EL KIND Y ADEMAS SI NO ESPECIFICO CUAL SIEMORE TOMARA EL QUE SE ACOMODE MEJOR TAMBIEN PUEDO ESPECIFICAR LO DE ESTILOS COMO LA PALETA ETC...

# recordar que el ; del final es lo mismo que el plt.show()


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head()
```

Uso de .kde y .displot para graficar distribuciones

>Gráfico para distribuciones y que se acumulen cumulative


```python
sns.histplot(data = tips, x= 'tip', bins = 15, cumulative=True);
```

Recordar que el hue nos clasifica segun el dato que le pasemos al hue

```python
sns.histplot(data = tips, x= 'tip', bins = 15,hue='sex');
```

# Stat

>ES LOQUE SE MOSTRARA EN EL EJE Y BASICAMENTE LO QUE VAMOS A TRABAJAR EN EL EJE Y

```python
sns.histplot(data = tips, x= 'tip', bins = 15,hue='sex',stat='frequency');
```

```python
#ahora con el porcent o probability , density ,e ldefault es count  y muchos mas
sns.histplot(data = tips, x= 'tip', bins = 15,hue='sex',stat='probability');
```

# MULTIPLE

>COMO SE PUEDE VER EXISTEN DOS VARIABLES ENTONCES YO PUEDE DECIRLE CON ESTA QUE ME PRESENTE DE CIERTA MANERA
> > STACK: CUANDO TERMINE UNO QUE INICIE EL OTRO
> > LAYER: LA DE DEFECTO
> > DODGE: ES QUE ME PRESENTE UNO AL LADO DEL  OTRO EL MEJOR
> > FILL: ES EL LLENAR

```python
sns.histplot(data = tips, x= 'tip', bins = 15,hue='sex',stat='count',multiple='stack');
```

```python
sns.histplot(data = tips, x= 'tip', bins = 10,hue='sex',stat='count',multiple='dodge');
```

```python
sns.histplot(data = tips, x= 'tip', bins = 15,hue='sex',stat='count',multiple='fill');
```

# Uso de .kde



>Gráfica de densidad

```python
sns.kdeplot(data=tips,x='tip',hue='sex');
```

> Con acumulación

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True);
```

>Crea una área debajo de la curva

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True);
```

>bw_adjust=1 ajustar la grafica y los ejes

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True, bw_adjust=1);
```

# OTRO TIPO DE GRAFICA ES LA exdf no es muy utilizado pero sirve para ciertos casos

>ESTE ES DE PROPORCION DE USO
>stat: lo de count osea lo mismo que el histograma etc...

```python
sns.ecdfplot(data=tips, x='tip', hue='sex');
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex',stat='count');
```

# DIAGRAMAS DE DISTRIBUCION (ES LO QUE NOS FACILITA SEABORN) solo es pasarle el tipo de grafico y ya no necesitamos recordar los nombres de los anteripres aso que es mejor usar displot

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde',multiple='stack');
```

```python
sns.displot(data=tips,x='tip',hue='smoker',multiple='dodge')
```
