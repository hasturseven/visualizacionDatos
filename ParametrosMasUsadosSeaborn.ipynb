# Parámetros más comunes al usar Seaborn Cada uno de los gráficos utiliza ciertos parámetros que tu podrás gestionar.

```python
import seaborn as sns
import matplotlib.pyplot as plt
#------------ CREANDO GRAFICO DE BARRAS CON SEABORN --------------
tips = sns.load_dataset('tips') #---> Dataset de seaborn
tips
```

# Puedes usar ’ ; ’ al final de la linea de código para remplazar el 'plt.show()'

```python
sns.displot(data=tips,x='total_bill')
plt.show()
```

>ahora usando eje y como el de arriba genero fue un histograma y le daremos son dos ejes el mismo me dira que no funciona bien esa grafica asi que me recomendara usar otra no me dira si no que se vera mal

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

el valor hill me dara como un deglose segun la varaible osea separandolo por colores etc

```python
sns.displot(data=tips,x='total_bill',hue='sex')
plt.show()
```

RECORDANDO QUE EL DISPLOT ES GRAFICO DE DISTRIBUCION Y SI USO KIND LE DIGO QUE TIPO DE GRAFICO TRABAJARE PARA ENTENDERMAS MIRARR LA IMAGEN INFORMATIVA QUE TENGO y el legend es que aparezca las etiquetas en este caso segun el color es hombre y mujer

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=True)
plt.show()
```

TAMBIEN PODEMOS USAR LA PALETA DE COLORES

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=True,palette='Set2')
plt.show()
```

tambien podemos manejar la tranparencia con alpha

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='ecdf',legend=True,palette='Set2',alpha=0.5)
plt.tick_params()
plt.show()
```
